```python
def read_data(file_list,label) :
    
    print(file_list,label)
    
    gfile = tf.io.read_file(tf.strings.join(['train/',file_list]))
    image = tf.io.decode_image(gfile)
    
    
    label = tf.cast(label, tf.uint8)
    return image, label
```

```python
# datasets ë§Œë“¤ê¸°
train_ds = tf.data.Dataset.from_tensor_slices((file_list, label_list))
train_ds = train_ds.map(read_data)
train_ds = train_ds.shuffle(1000)
train_ds = train_ds.batch(32)
train_ds = train_ds.repeat()
```

```python
val_ds = tf.data.Dataset.from_tensor_slices((val_list, val_label_list))
val_ds = test_ds.map(read_data)
val_ds = test_ds.batch(32)
val_ds = test_ds.repeat()
```

```python
input_shape = (32,32,3)
batch_size = 32
```

```python
inputs = layers.Input(input_shape)


net = layers.Conv2D(32, 3, 1, 'same')(inputs)
net = layers.Activation('relu')(net)
net = layers.Conv2D(32, 3, 1 ,'same')(net)
net = layers.Activation('relu')(net)
net = layers.MaxPool2D(2)(net)
net = layers.Dropout(0.25)(net)


net = layers.Conv2D(32, 3, 1, 'same')(inputs)
net = layers.Activation('relu')(net)
net = layers.Conv2D(32, 3, 1 ,'same')(net)
net = layers.Activation('relu')(net)
net = layers.MaxPool2D(2)(net)
net = layers.Dropout(0.25)(net)


net = layers.Flatten()(net)
net = layers.Dense(512)(net)
net = layers.Activation('relu')(net)
net = layers.Dense(1)(net)
net = layers.Activation('sigmoid')(net)
```

```python
model = tf.keras.Model(inputs=inputs, outputs=net, name='mnist')
```

```python
model.compile(
    loss = tf.keras.losses.binary_crossentropy,
    optimizer = tf.keras.optimizers.Adam(),
    metrics = ['accuracy']
)
```

```python
#len(data_paths), data_paths[0] , len(test_paths), test_paths[0]

num_dataset = 15000 
steps_per_epoch = num_dataset // batch_size

num_testset = 2500
validation_steps = num_testset // batch_size

steps_per_epoch, validation_steps
```
