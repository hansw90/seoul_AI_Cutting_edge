```python
image_pil = Image.open('cat2.jpg')
plt.imshow(image_pil)
```

```python
f_image = np.flip(image_pil,0)
plt.imshow(f_image)
```

```python
f_image = np.flip(image_pil,1)
plt.imshow(f_image)
```

```python
def rand_flip(image) :
    if np.random.randint(2) :
        f_image = np.flip(image_pil,1)
        
    if np.random.randint(2) :
        f_image = np.flip(image_pil,0)
        
    return f_image
```

```python
for _ in range(5) :
    image = rand_flip(image_pil)
    plt.imshow(image)
    plt.show()
```

```python
image = np.rot90(image_pil)
plt.imshow(image)
```

```python
# width/height shift
image = np.array(image_pil)
x,y,c = image.shape

# dtpye을 정확하게 해주지 않으면 제대로 복사를 못한다.
zero_image = np.zeros((x,y,c),dtype=np.int32)
zero_image.shape
```

```python
zero_image[0:y-100, 0:x-200] = image[100:, 200:]
```

```python
plt.imshow(zero_image)
```

```python
# zoom

zoom_image =  image[500:1000,100:500]

plt.imshow(zoom_image)
```
